# Lesson 2: Addition of System of Coplanar Forces

## Coplanar Force Resultants

To determine the resultant of several **coplanar forces**, i.e., forces that all lie in the same plane, each force is resolved into its $x$ and $y$ components, and then the respective components are added using *scalar algebra* since they are collinear. The resultant force is then formed by adding the *resultant components* using the parallelogram law.

**Vector Notation**

$$\mathbf{F_1} = F_{1x}\mathbf{i} + F_{1y}\mathbf{j}\\
\mathbf{F_2} = F_{2x}\mathbf{i} + F_{2y}\mathbf{j}\\
\mathbf{F_3} = F_{3x}\mathbf{i} + F_{3y}\mathbf{j}$$

The vector resulant is:

$$\mathbf{R} = \mathbf{F_1} + \mathbf{F_2} + \mathbf{F_3}\\
\mathbf{R} = F_{1x}\mathbf{i} + F_{1y}\mathbf{j} - F_{2x}\mathbf{i} + F_{2y}\mathbf{j} + F_{3x}\mathbf{j} - F_{3y}\mathbf{j}\\
\mathbf{R} = (F_{1x}-F_{2x}+F_{3x})\mathbf{i} + (F_{1y}+F_{2y}-F_{3y})\mathbf{j}\\
\mathbf{R} = (R_x)\mathbf{i} + (R_y)\mathbf{j}$$

**Scalar Notation**

$$R_x = F_{1x} - F_{2x} + F_{3x}\\
R_y = F_{1y} + F_{2y} - F_{3y}$$

In general, the components of the resultant force of any number of coplanar forces can be represented by the algebraic sum of the $x$ and $y$ components of all the forces.

$$R_x = \sum{F_x}\\
R_y = \sum{F_y}$$

$$R = \sqrt{R_x^2 + R_y^2}$$

$$\theta = \tan^{-1} \frac{R_y}{R_x}$$
 

## Example 1

The end of the boom *O* in the figure is subjected to the three concurrent and coplanar forces. Determine the magnitude and direction of the resultant force.



**Solution**

```python
from math import sin, cos, pi, atan, sqrt
import handcalcs.render
```

```python
%%render params

F_1 = 400 # All forces are in N
F_2 = 250
F_3 = 200
```

```python
%%render

theta_1 = 45*pi/180 # all angles are in radians
theta_2 = atan(3/4)
```

**Calculation of x-component of the resultant**

```python
%%render

R_x = -F_1+F_2*sin(theta_1)-F_3*cos(theta_2) # N
```

**Calculation of y-component of the resultant**

```python
%%render

R_y = F_2*cos(theta_1)+ F_3*sin(theta_2) # N
```

**Calculation of the magnitude of the resultant, R**

```python
%%render

R = sqrt(R_x**2+R_y**2) # N
```

**Calculation of the direction of the resultant**

```python
%%render

theta = atan(R_y/R_x)*180/pi  # angle in degrees
```

## Example 2

Determine the magnitude of the resultant force and its direction $\theta$, measured counterclockwise from the positive $x$ axis.



**Solution**

```python
%%render params

F_1 = 15 # all forces are in kN
F_2 = 20
F_3 = 15
```

```python
%%render

theta_1 = atan(3/4) # all angles are in radians
theta_2 = pi/2
theta_3 = atan(3/4)
```

**Calculation of $R_x$**

```python
%%render

R_x = F_1*cos(theta_1)+F_2*cos(theta_2)+F_3*cos(theta_3) # kN
```

**Calculation of $R_y$**

```python
%%render

R_y = -F_1*sin(theta_1)+F_2*sin(theta_2)+F_3*sin(theta_3) # kN
```

**Calculation of $R$**

```python
%%render

R = sqrt(R_x**2+R_y**2) # kN
```

**Direction $\theta$ measured counterclockwise from the positive x-axis**

```python
%%render

theta = atan(R_y/R_x)*180/pi # acute angle in degrees
```
