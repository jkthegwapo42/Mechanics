# Lesson 5: Moment of a Force - Scalar Formulation

- The **Moment** of force is measure of its tendency to cause a body to rotate about a specific point or axis. 
- In order for a moment to develop, the force must act upon a body in such a manner that the body would begin to twist. 
- This occurs every time a force is applied so that it does not pass through the centroid of the body.
- A moment is due to a force not having an equal and opposite force directly along it's lind of action.

## Magnitude



The magnitude of $\mathbf{M_O}$ is:

$$M_O = Fd$$

where *d* is the **moment arm** or *perpendicular distance* from the axis at *O* to the line of action of the force. Units of moment magnitude consists of force times distance, e.g., $N\cdot m$.

## Direction

- The direction of $\mathbf{M_O}$ is defined by its **moment axis** which is perpendicular to the plane that contains the force $\mathbf{F}$ and its moment arm *d*.
- The right-hand rule is used to establish the sense of direction of $\mathbf{M_O}$ where the natural curl of the fingers of the right hand, as they are drawn towards the palm, represents the rotation, or the tendency for rotation caused by the moment. The thumn of the right hand will give the directional sense of $\mathbf{M_O}$.
- In two dimensions, counterclockwise rotation produces a positive value.




## Resultant Moment

- For two dimensional problems, where all the forces lie within the $x-y$ plane, the resultant moment $(\mathbf{M_R})_O$ about point *O* (the z-axis) can be determined by *finding the algebraic sum* of the moments caused by all the forces in the system.
- As a convention, *counterclockwise moments* are considered positve and *clockwise moments* are negative.



$$(\mathbf{M_R})_O = \sum Fd$$
