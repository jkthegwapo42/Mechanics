# Lesson 7: Moment of a Force - Vector Formulation

The moment of a force $\mathbf{F}$ about point *O* can be expressed using the vector product,

$$\mathbf{M_O} =\mathbf{r}\times\mathbf{F}$$

where $\mathbf{r}$ is a position vector directed from *O* to any point on the line of action of $\mathbf{F}$.



## Magnitude

The magnitude of the cross product is defined from $M_o = rF\sin{\theta}$, where the angle $\theta$ is measured between the tails of $\mathbf{r}$ and $\mathbf{F}$.


$$M_o =rF\sin{\theta} = F(r\sin{\theta}) = Fd$$



## Direction

The direction and sense of $\mathbf{M_O}$ are determined by the right-hand rule as it applies to the cross product.

## Principle of Transmissibility

The principle of transmissibility allows $\mathbf{F}$ can be applied at any point along its line of action and still create the **same moment** about point *O*.



$$\mathbf{M_O} = \mathbf{r_1}\times\mathbf{F} = \mathbf{r_2}\times\mathbf{F} = \mathbf{r_3}\times\mathbf{F}$$

## Cartesian Vector Formulation

$$\mathbf{M_O} = \mathbf{r}\times\mathbf{F} = \left|\begin{array}{ccc}\mathbf{i} &\mathbf{j} &\mathbf{k}\\
r_x & r_y & r_z\\
F_x & F_y & F_z \end{array}\right|$$

where $r_x,r_y,r_z$ represents the $x,y,z$ components of the position vector drawn from point *O* to **any point** on the line of action of the force

$F_x,F_y,F_z$ represents the$x,y,z$ components of the force vector.



## Resultant Moment of a System of Forces

If a body is acted upon by a system of forces, the resultant moment of the forces about point *O* can be determined by vector addition of the moment of each force.



$$(\mathbf{M_R}_O = \sum(\mathbf{r}\times\mathbf{F})$$

## Example 1

Determine the moment produced by the force $\mathbf{F}$ in the figure about point *O*. Express the result as a Cartesian vector.



**Solution**

Either $\mathbf{r_A}$ or $\mathbf{r_B}$ can be used to determine the moment about point O.

```python
from numpy import array, cross
from math import sqrt
```

Determination of vector $\mathbf{AB}$.


```python
## Coordinates of point A

x_A = 0
y_A = 0
z_A = 12

## Coordinates of point B

x_B = 4
y_B = 12
z_B = 0

## Vector AB

x = x_B-x_A
y = y_B-y_A
z = z_B-z_A

AB = array([x,y,z])
AB
```

Determination of unit vector $\mathbf{u}_{AB}$

```python
## Magnitude of vector AB

AB_mag = sqrt(x**2+y**2+z**2)
AB_mag

## Unit vector AB

u_AB = AB/AB_mag
u_AB
```

Expressing $\mathbf{F}$ as a Cartesian vector

```python
F_mag = 2; # force in kN

F =F_mag*u_AB; # force F expressed as a Cartesian vector
F
```

Using $\mathbf{r}_A$

```python
# Position vector r_A

r_A = array([x_A,y_A,z_A])
r_A
```

Determination of $\mathbf{M}_O$

```python
# Moment M_O in kN-m using r_A

M_O = cross(r_A,F)
M_O
```

Using $\mathbf{r}_B$

```python
# Position vector r_B

r_B =array([x_B,y_B,z_B])
r_B
```

Determination of $\mathbf{M}_O$

```python
# Moment M_O in kN-m using r_B

M_O =cross(r_B,F)
M_O
```

Ans

$$\mathbf{M}_O = -16.52\mathbf{i} +5.51\mathbf{j}$$

## Example 2

Two forces act on the rod shown in the figure below. Determine the resultant moment they create about the flange at *O*. Express the result as a Cartesian vector.



**Solution**

```python
# Given
# All forces are in N

F_1 = array([-600,400,200])
F_2 =array([800,400,-300])
```

Determination of position vectors $\mathbf{r}_A$ and $\mathbf{r}_B$



```python
# Coordinates of A
# All distances are in meters

x_A = 0
y_A = 2.5
z_A = 0

# Position vector r_A

r_A = array([x_A,y_A,z_A])
r_A
```

```python
# Coordinates of B
# All distances are in meters

x_B = 2
y_B = 2.5
z_B = -1

# Positiob vector r_B

r_B =array([x_B,y_B,z_B])
r_B
```

Calculation of resultant moment $\mathbf{M}_O$

```python
M_O = cross(r_A,F_1)+cross(r_B,F_2)
M_O
```

Ans.

$$\mathbf{M}_O = 150\mathbf{i}-200\mathbf{j}+300\mathbf{k}$$
